```python
import glob
import json
import matplotlib.pyplot as plt
from collections import Counter
```

```python
class Trace:
    def __init__(self, filename):
        self.filename = filename
        with open(filename) as f:
            self.data = json.load(f)
        self.loops = max([int(k) for k in self.data.keys()])
        self.game_result = self.data[str(self.loops)]['game_result']
        self.map = self.data[str(self.loops)]['map']
        self.model = self.data[str(self.loops)]['model']
```

```python
trace_files = glob.glob('logs/Qwen2.5-72B-Instruct/*/trace.json')
traces = []
for trace_file in trace_files:
    with open(trace_file) as f:
        trace = json.load(f)
        last_round = str(max([int(key) for key in trace.keys()]))
        if "game_result" in trace[last_round]:
            traces.append(Trace(trace_file))
            if trace[last_round]["game_result"] == "Victory":
                print(trace_file)
```

```python
result_counter = Counter()
for trace in traces:
    result_counter[trace.game_result] += 1

print(result_counter.most_common())
```

```python
for key in ["Victory", "Defeat"]:
    average_loops = sum([trace.loops for trace in traces if trace.game_result == key]) / result_counter[key]
    average_time = average_loops / 4 / 100 * 17 / 60
    average_actions = sum([len(trace.data) - 1 for trace in traces if trace.game_result == key]) / result_counter[key]
    print(f"[{key}]")
    print(f"Average loops: {int(average_loops)}")
    print(f"Average time: {round(average_time, 2)} minutes")
    print(f"Average actions: {round(average_actions, 2)}")
print("Total actions:", sum([len(trace.data) - 1 for trace in traces]))
```
